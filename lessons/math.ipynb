### **Funções Matemáticas**


##### Função Linear


```python
def funcao(x):
    y = 2 * x + 1
    return y
```

```python
valores_x = range(1, 6)
valores_y = []

for x in valores_x:
    y = funcao(x)
    print(f"f(x) = {y}")
```

```python
x = symbols("x")
y = 2 * x + 1
plot(y, title="f(x) = 2x + 1")
```

##### Função Quadrática


```python
def funcao(x):
    y = x**2 + 1
    return y
```

```python
valores_x = range(1, 6)
valores_y = []

for x in valores_x:
    y = funcao(x)
    print(f"f(x) = {y}")
```

```python
x = symbols("x")
y = x**2 + 1
plot(y, title="f(x) = x^2 + 1")
```

##### _Função Tridimensional_


```python
def funcao(x, y):
    z = 2 * x + 3 * y
    return z
```

```python
valores_x = range(1, 6)
valores_y = range(6, 11)
valores_z = []


for i in range(5):
    z = funcao(valores_x[i], valores_y[i])
    valores_z.append(z)


for z in valores_z:
    print(f"f(x, y) = {z}")
```

```python
x, y = symbols("x y")
z = 2 * x + 3 * y
plot3d(z, title="f(x, y) = 2x + 3y")
```

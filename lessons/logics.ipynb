# **Lógica Proposicional**


```python
import numpy as np
import pandas as pd
```

```python
valor1 = np.array([True, True, False, False])
valor2 = np.array([True, False, True, False])
```

```python
conjuncao = valor1 & valor2
disjuncao1 = valor1 | valor2
disjuncao2 = valor1 ^ valor2
condicao = ~valor1 | valor2
bicondicao = valor1 == valor2
```

### **Conjunção (E)**


```python
pd.DataFrame(
    {
        "Valor P": valor1,
        "Valor Q": valor2,
        "Conjunção (P ^ Q)": conjuncao,
    }
)
```

### **Disjunção (OU)**


```python
pd.DataFrame(
    {
        "Valor P": valor1,
        "Valor Q": valor2,
        "Disjunção (P v Q)": disjuncao1,
    }
)
```

### **Disjunção (OU... OU)**


```python
pd.DataFrame(
    {
        "Valor P": valor1,
        "Valor Q": valor2,
        "Disjunção (P ⊻ Q)": disjuncao2,
    }
)
```

### **Condicional (SE... ENTÃO)**


```python
pd.DataFrame(
    {
        "Valor P": valor1,
        "Valor Q": valor2,
        "Condição (P -> Q)": condicao,
    }
)
```

### **Bicondicional (SE E SOMENTE SE)**


```python
pd.DataFrame(
    {
        "Valor P": valor1,
        "Valor Q": valor2,
        "Bicondição (P <-> Q)": bicondicao,
    }
)
```
